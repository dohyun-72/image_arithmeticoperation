```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# 이미지 두장 불러오기
img1 = cv2.imread('imgset1.png', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('imgset2.png', cv2.IMREAD_GRAYSCALE)

# 크기 조정
if img1.shape != img2.shape:
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# 이미지 더하기 연산
added_result = cv2.add(img1, img2)

# 결과 출력
print("image add")
combined_add = np.hstack((img1, img2, added_result))
cv2_imshow(combined_add)
```

```python
# img1에서 img2 빼기
sub1 = cv2.subtract(img1, img2)

# img2에서 img1 빼기
sub2 = cv2.subtract(img2, img1)

# 결과 출력
print("subtraction 연산 결과 비교")
combined_sub_all = np.hstack((img1, img2, sub1, sub2))
cv2_imshow(combined_sub_all)
```

```python
# 이미지 곱하기
# 두 이미지의 값을 곱하면 각각의 수치가 배로 뛰기때문에 금방 255를 넘기게되고, 흰색으로 나타납니다.
mul_result = cv2.multiply(img1, img2)

# 2. 이미지 나누기
# 이미지가 거의 유사하기 때문에 값이 확 낮아지며 검정색 단색 이미지로 보입니다.
div_result = cv2.divide(img1, img2)

print("원본 사진과 multiplication, division 결과 비교")
combined_mul_div = np.hstack((img1, img2, mul_result, div_result))
cv2_imshow(combined_mul_div)
```
